```python
import os
import pandas as pd
```

```python
# 경로는 환경에 맞게 수정. 예: 환경 변수 DATA_ROOT 또는 로컬 경로
folder_path = os.environ.get("DATA_ROOT", "./resources/archive")
dataset_path = os.path.join(folder_path, "lfw-deepfunneled", "lfw-deepfunneled")
```

```python
csv_path = os.path.join(folder_path, "people.csv")
csv_path
```

```python
df = pd.read_csv(csv_path)
df = df[~df['images'].isna()].reset_index(drop=True)
df['images'] = df['images'].astype(int)
df
```

```python
valid_df = df[df['images']>1].reset_index(drop=True)
valid_df
```

```python
valid_df.to_csv('data_for_face_reco.csv', index=False)
```
